```python
a = eval(input("Enter a number "))
b = eval(input("Enter a number "))
ops = eval(input("Enter 1 for addtion 2 for substraction 3 for multiplication and 4 for division :"))
def add():
    print("The addiotn of {} and {} is {}".format(a,b,a+b))
def sub():
    print("The substraction of {} and {} is {}".format(a,b,a-b))
def mul():
    print("The multiplication of {} and {} is {}".format(a,b,a*b))
def div():
    print("The divisn of {} and {} is {}".format(a,b,round(a/b,2)))
def main():
    if ops==1: 
        add()
    elif ops==2: 
        sub()
    elif ops==3: 
        mul()
    elif ops==4: 
        div()
    else:
        print("Enter number between 1 to 4")
main()
```
